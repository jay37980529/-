```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt 
import numpy as np
from pandas import Series,DataFrame
```

```python
data1 = pd.read_csv('2018.csv')
data1
```

```python
data1['Districts2'] = ""
```

```python
data1
```

```python
data1.rename(columns = {'discricts' : 'District'}, inplace = True)
```

```python
data1
```

```python
for index, row in data1.iterrows():
    if '[' and ']' in row['District']:
        start_index = row['District'].find('[')
        end_index = row['District'].find(']')
        data1.loc[index, 'Districts2'] = data1.loc[index]['District'][start_index+1: end_index]
        
    else:
        data1.loc[index, 'Districts2'] = data1.loc[index]['District']
```

```python
data1
```

```python
nep_districts = gpd.read_file(r'TWN_adm/TWN_adm2.shp')
nep_districts
```

```python
nep_districts = nep_districts[['NAME_2', 'geometry']]
nep_districts.rename(columns = {'NAME_2' : 'District'}, inplace = True)
```

```python
nep_districts
```

```python
for index, row in nep_districts['District'].iteritems():
    if row in data1['District'].tolist():
        pass
    else:
        print('The district ', row , ' is NOT in the data1 list')
```

```python
nep_districts.replace('Changhwa', '彰化縣', inplace = True)
nep_districts.replace('Chiayi', '嘉義市', inplace = True)
nep_districts.replace('Hsinchu', '新竹市', inplace = True)
nep_districts.replace('Hualien', '花蓮縣', inplace = True)
nep_districts.replace('Ilan', '宜蘭縣', inplace = True)
nep_districts.replace('Kaohsiung', '高雄市', inplace = True)
nep_districts.replace('Kaohsiung City', '高雄市', inplace = True)
nep_districts.replace('Keelung City', '基隆市', inplace = True)
nep_districts.replace('Miaoli', '苗栗縣', inplace = True)
nep_districts.replace('Nantou', '南投縣', inplace = True)
nep_districts.replace('Penghu', '澎湖縣', inplace = True)
nep_districts.replace('Pingtung', '屏東縣', inplace = True)
nep_districts.replace('Taichung', '臺中市', inplace = True)
nep_districts.replace('Taichung City', '臺中市', inplace = True)
nep_districts.replace('Taipei City', '臺北市', inplace = True)
nep_districts.replace('Tainan', '臺南市', inplace = True)
nep_districts.replace('Tainan City', '臺南市', inplace = True)
nep_districts.replace('Taipei', '臺北市', inplace = True)
nep_districts.replace('Taitung', '臺東縣', inplace = True)
nep_districts.replace('Taoyuan', '桃園市', inplace = True)
nep_districts.replace('Yunlin', '雲林縣', inplace = True)
```

```python
nep_districts
```

```python
data1
```

```python
nep_districts['area'] = nep_districts.area*10000
```

```python
nep_districts = nep_districts.merge(data1, on = 'District')
nep_districts
```

```python
nep_districts['amo_den (amount/sq. km)'] = nep_districts['amount']/100000
```
